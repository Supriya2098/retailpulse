```python
import os

os.makedirs(

    ".github/workflows",

    exist_ok=True

)

print("✅ GitHub workflow folders created")
```

```python
%%writefile .github/workflows/ci-cd.yml

name: RetailPulse CI/CD Pipeline

on:

  push:

    branches:
      - main

  pull_request:

    branches:
      - main

jobs:

  build-and-test:

    runs-on: ubuntu-latest

    steps:

    # ========================================================
    # CHECKOUT
    # ========================================================

    - name: Checkout Repository

      uses: actions/checkout@v4

    # ========================================================
    # PYTHON SETUP
    # ========================================================

    - name: Setup Python

      uses: actions/setup-python@v5

      with:
        python-version: '3.11'

    # ========================================================
    # INSTALL DEPENDENCIES
    # ========================================================

    - name: Install Dependencies

      run: |

        python -m pip install --upgrade pip

        pip install -r requirements.txt

    # ========================================================
    # RUN TESTS
    # ========================================================

    - name: Run Basic Validation

      run: |

        python --version

        echo "RetailPulse CI Passed ✅"

    # ========================================================
    # BUILD DOCKER IMAGE
    # ========================================================

    - name: Build Docker Image

      run: |

        docker build -t retailpulse .

    # ========================================================
    # SUCCESS
    # ========================================================

    - name: Deployment Ready

      run: |

        echo "🚀 RetailPulse ready for deployment"
```
